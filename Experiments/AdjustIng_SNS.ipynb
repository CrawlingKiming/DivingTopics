## 문자 데이터 전처리 

필요 Data = SNS_ 문서들 
Output = processed_SNS 


이곳은 실제적인 문자 데이터 전처리를 담당하는 코드이다. 
전처리 및 불필요한 주석, 문자 포맷이 이곳에서 모두 제거되며, 
시간이 많이 걸리는 형태소 분석을 끝내고 파일에 저장된다.


```python
from collections import defaultdict
import re 
from eunjeon import Mecab

stop_text = open("KoreanStopwords.txt", "r", encoding = "utf-8")
stop_list = []
for line in stop_text :
    line = line[:-1]
    stop_list.append(line)
stop_text.close()
stop_list.extend(["\xa0","\u200b"])
```

```python
mecab = Mecab()
```

```python
delet_re = re.compile(r'[:/(){}\[\]\|@,;\\]+\w+|[:/(){}\[\]\|@,;\\]+|[0-9]+')
```

```python
i = 0
j = 0
```

```python
def tokenizing(text):
    global i 
    global j
    i += 1 
    if i % 1000 == 0 :
        print("{0}th iteration, time {1}".format(i, time.time()))
    #returns 형태소, exclude surplus tags.
    try:
        text = str(text)
        text = delet_re.sub('', text)
        tok = mecab.morphs(text)
        
    except UnicodeDecodeError:
        j += 1 
        print("{0}th text had been dumped..".format(j))
    except ValueError :
        j += 1
        print(text)
        print("{0}th text had been dumped..".format(j))
        
        return []
   
    return tok
    

def pre_processing(text) :
    result = []
        
    for token in tokenizing(text):
        if token not in stop_list :
            result.append(token)
        
    return result

def prepare_process(text):
    result = []
        
    for token in tokenizing(text):
        if token not in stop_list :
            result.append(token)
    
    result = " ".join(result)
    return result

def read_token(text):
    result = text.split()
    
    return result
```

```python
def FilterData(fname, i):   
    df = pd.read_excel(fname)
    df["DATE"] = df["DATE"] // 10**6
    or_shape = df.shape
    #df = df.loc[(20180401<=df["DATE"]) & (df["DATE"] <= 20181124) , :]
    df["CONTENT"]= df['CONTENT'].map(prepare_process) 
    in_shape = df.shape
    df.dropna( how = "any")
    na_shape = df.shape
    new_fname = "processed_SNS_{0}.xlsx".format(i)
    df.to_excel(new_fname, index = False, engine = "xlsxwriter")
    print("{0} is done.".format(fname))
    print("Original shape : {0}, Date_Indexed Shape : {1}, Dropped NA shape : {2}".format(or_shape, in_shape, na_shape))
    
```
