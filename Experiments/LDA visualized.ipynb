# LDA 시각화 
LDA Topic Model은 시각화 하기 난해하다. 
pyLDAvis를 이용하여 시각화한 코드이다. 

```python
import gensim
```

```python
# load lda model and bow_corpus
lda_trained = gensim.models.LdaModel.load("lda_trained.model")
```

```python
with open("bow_corpus.pickle", "rb") as f : 
    bow_corpus = pickle.load(f)
```

```python
with open("lda_trained.model.id2word", "rb") as f : 
    id2word = pickle.load(f)
```

```python
import pyLDAvis.gensim
```

```python
import pyLDAvis.gensim
pyLDAvis.enable_notebook()
vis = pyLDAvis.gensim.prepare(lda_trained, bow_corpus, id2word)
pyLDAvis.display(vis)
```
